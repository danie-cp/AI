Find the k-th level friend of a given user

```python
from collections import deque

# Start the graph search from root using a doubly-ended queue

def FindFriends(G, root, level):

    # deque is list-like container with fast appends and pops on either end
    visited, queue = set(), deque([root])
    visited.add(root)

    if root not in G:                            #Checks if the person is in the graph
      print("No friends")
      return

    k= 0                                         #the number of parent nodes + their first-level children
    num_iter =0                                  #the number of level
    i =0                                         #the hypothetical iteration for the visited queue



      ####################################################################################################
      # We are interested when the queue set only contains the children of the nodes in the visited set.
      # To keep track of this occurance, as i iterates, k contains the number of parent nodes and their children.
      # When k and i are equal, this is when the new children are in the queue.
      # num_iter tracks the amount of times this occur, thus its level.
      ##########################################################################################################


    # Keep exploring as long as the queue is not empty
    while queue:

        if (i ==k):
          k = k + len(queue)
          num_iter = num_iter+1

          if (num_iter == level+1):              #Once, the desired lvl is reached, return the friend(s)
            print(queue)
        i = i+1


        # Dequeue a vertex from queue
        vertex = queue.popleft()


        # If not visited, mark it as visited, and enqueue it
        for neighbor in G[vertex]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
```

```python

G = {
  'Richard' : ['Bob'],
  'Bob' : ['Richard','Rob','Pam'],
  'Rob' : ['Bob'],
  'Pam' : ['Bob','Roger', 'Peter'],
  'Roger' : ['Pam','Anna'],
  'Anna' : ['Roger'],
  'Peter' : ['Pam','Amy'],
  'Amy' : ['Peter']
}


FindFriends(G, "Bob", 3)
```
